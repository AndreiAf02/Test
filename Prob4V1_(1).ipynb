```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
mu_x1=1
mu_y1 = -1

mu_x2=0
mu_y2 = 2

mu_x3=-1
mu_y3 = 0

sigma=0.75

a=2
b=3
```

```python
Data1 = pd.DataFrame()
Data2 = pd.DataFrame()
Data3 = pd.DataFrame()

np.random.seed(1)
Data1['x']=np.random.normal(mu_x1, sigma, 20)
Data1['y']=np.random.normal(mu_y1, sigma, 20)
Data1['Class']=1

Data2['x']=np.random.normal(mu_x2, sigma, 20)
Data2['y']=np.random.normal(mu_y2, sigma, 20)
Data2['Class']=2

Data3['x']=np.random.normal(mu_x3, sigma, 20)
Data3['y']=np.random.normal(mu_y3, sigma, 20)
Data3['Class']=3

Data_all = pd.concat([Data1, Data2, Data3], axis=0)
Data = Data_all.drop(['Class'], axis=1)
```

```python
plt.figure(figsize=(8, 6), dpi=1200)
plt.plot(Data1['x'], Data1['y'], 'b.', label='1')
plt.plot(Data2['x'], Data2['y'], 'g.', label='2')
plt.plot(Data3['x'], Data3['y'], 'r.', label='3')
plt.ylabel('y')
plt.xlabel('x')
plt.title('True Data Distribution')
plt.legend()
```

```python
## Adapted from ISLP textbook

kmeans = KMeans(n_clusters=3, random_state=3, n_init=20).fit(Data)
```

```python
Data_all['K_means']=kmeans.labels_
Data_all
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8, 6), dpi=1200)
plt.scatter(Data['x'], Data['y'], c=kmeans.labels_)
plt.show()
```

**Point D:**

```python
kmeans = KMeans(n_clusters=4, random_state=4, n_init='auto').fit(Data)
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8, 6), dpi=1200)
plt.scatter(Data['x'], Data['y'], c=kmeans.labels_)
plt.show()
```

```python
mu_1 = 1
mu_2 = 0
mu_3 = -1
sigma=2
```

```python
Data1 = pd.DataFrame()
Data2 = pd.DataFrame()
Data3 = pd.DataFrame()

np.random.seed(1)

for i in range(1,51):
  # print(i)
  Data1['Class']=1
  Data1['d'+str(i)]=np.random.normal(mu_1, sigma, 20)

  Data2['Class']=2
  Data2['d'+str(i)]=np.random.normal(mu_2, sigma, 20)

  Data3['Class']=3
  Data3['d'+str(i)]=np.random.normal(mu_3, sigma, 20)

Data_all = pd.concat([Data1, Data2, Data3], axis=0)
Data = Data_all.drop(['Class'], axis=1)
```

```python
kmeans = KMeans(n_clusters=3, random_state=3, n_init='auto').fit(Data)
```

```python
Data_all['K_means']=kmeans.labels_
Data_all[['Class', 'K_means']]
```

```python
kmeans = KMeans(n_clusters=4, random_state=4, n_init='auto').fit(Data)

Data_all['K_means']=kmeans.labels_
Data_all[['Class', 'K_means']]
```
